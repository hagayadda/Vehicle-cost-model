### First we turned the "undefined" value into a missing value and then completed values ​​according to the most common value in similar cars

```python
# Replace "לא מוגדר" with NaN
data_car['Gear'].replace("לא מוגדר", pd.NA, inplace=True)

# Fill missing values in 'Gear' by the most frequent value of similar cars grouped by 'manufactor', 'Year', and 'model'
data_car['Gear'] = data_car.groupby(['manufactor', 'Year', 'model'])['Gear'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))

# Check for remaining missing values in the 'Gear' column
missing_gear = data_car['Gear'].isnull().sum()
print(f"Missing values in 'Gear' after first step: {missing_gear}")

# If there are still missing values, fill them by the most frequent value of similar cars grouped by 'manufactor' and 'Year'
if missing_gear > 0:
    data_car['Gear'] = data_car.groupby(['manufactor', 'Year'])['Gear'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mode()))

    # Check for remaining missing values in the 'Gear' column
    missing_gear = data_car['Gear'].isnull().sum()
    print(f"Missing values in 'Gear' after second step: {missing_gear}")

```

### First we removed commas from the engine volume number to make the value a numerical value, then we made every number smaller than 500 a missing value because such small engines do not exist and it is probably a mistake

```python
data_car['capacity_Engine'] = data_car['capacity_Engine'].str.replace(',', '').astype(float)


# Identify and replace engine capacities that are less than 500 with NaN
data_car.loc[data_car['capacity_Engine'] < 500, 'capacity_Engine'] = pd.NA
```

### Completion of values ​​according to the average of similar vehicles in the same category

```python
# Fill missing values in 'capacity_Engine' by the mean value of similar cars grouped by 'manufactor', 'Year', and 'model'
data_car['capacity_Engine'] = data_car.groupby(['manufactor', 'Year', 'model'])['capacity_Engine'].transform(lambda x: x.fillna(x.mean()))

# Check for remaining missing values in the 'capacity_Engine' column
missing_capacity_engine = data_car['capacity_Engine'].isnull().sum()
print(f"Remaining missing values in 'capacity_Engine': {missing_capacity_engine}")

# If there are still missing values, fill them by the mean value of similar cars grouped by 'manufactor' and 'Year'
data_car['capacity_Engine'] = data_car.groupby(['manufactor', 'Year'])['capacity_Engine'].transform(lambda x: x.fillna(x.mean()))

# Check for remaining missing values in the 'capacity_Engine' column
missing_capacity_engine = data_car['capacity_Engine'].isnull().sum()
print(f"Remaining missing values in 'capacity_Engine': {missing_capacity_engine}")

# If there are still missing values, fill them by the mean value of similar cars grouped by 'manufactor'
data_car['capacity_Engine'] = data_car.groupby('manufactor')['capacity_Engine'].transform(lambda x: x.fillna(x.mean()))

# Check for remaining missing values in the 'capacity_Engine' column
missing_capacity_engine = data_car['capacity_Engine'].isnull().sum()
print(f"Remaining missing values in 'capacity_Engine': {missing_capacity_engine}")

# If there are still missing values, fill them by the overall mean of the column
data_car['capacity_Engine'].fillna(data_car['capacity_Engine'].mean(), inplace=True)

# Check for remaining missing values in the 'capacity_Engine' column
missing_capacity_engine = data_car['capacity_Engine'].isnull().sum()
print(f"Remaining missing values in 'capacity_Engine': {missing_capacity_engine}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, r2_score
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt

# Define the function to prepare the data
def prepare_data(data_car, threshold=0.9):
    # Add new features: Car Age and Km_per_Year
    data_car['Car_Age'] = 2024 - data_car['Year']
    data_car['Km_per_Year'] = data_car['Km'] / data_car['Car_Age']
    
    # Remove outliers in the Price column
    q_low = data_car['Price'].quantile(0.01)
    q_high = data_car['Price'].quantile(0.99)
    data_car = data_car[(data_car['Price'] > q_low) & (data_car['Price'] < q_high)]
    
    # Convert categorical variables to dummy variables (excluding 'Color' and 'City')
    data_car = pd.get_dummies(data_car, columns=['manufactor', 'model', 'Hand', 'Gear', 'Engine_type', 'Prev_ownership', 'Curr_ownership'])
    
    # Select relevant columns for the model
    relevant_columns = ['Year', 'capacity_Engine', 'Km', 'Car_Age', 'Km_per_Year', 'Price'] + \
                       [col for col in data_car.columns if col.startswith('manufactor_') or col.startswith('model_') or col.startswith('Hand_') or col.startswith('Gear_') or col.startswith('Engine_type_') or col.startswith('Prev_ownership_') or col.startswith('Curr_ownership_')]
    
    data_car_prepared = data_car[relevant_columns]
    
    # Separate features and target variable
    X = data_car_prepared.drop('Price', axis=1)
    y = data_car_prepared['Price']
    
    # Scale the features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Create a DataFrame with the scaled features
    X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
    
    # Calculate the correlation matrix
    corr_matrix = X_scaled_df.corr().abs()
    
    # Select upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    
    # Find features with correlation greater than the threshold
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    
    # Drop features with high correlation
    X_scaled_df = X_scaled_df.drop(columns=to_drop)
    
    return X_scaled_df, y

# Prepare the data
X, y = prepare_data(data_car, threshold=0.9)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Define parameter grid for Grid Search
param_grid_xgb = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 1.0]
}

# Perform Grid Search for XGBoost
xgb_model = xgb.XGBRegressor(random_state=0, objective='reg:squarederror')
grid_search_xgb = GridSearchCV(estimator=xgb_model, param_grid=param_grid_xgb, 
                               cv=10, n_jobs=-1, scoring='neg_mean_squared_error')
grid_search_xgb.fit(X_train, y_train)
best_xgb = grid_search_xgb.best_estimator_

# Evaluate the best XGBoost model on the test set
y_pred_xgb_best = best_xgb.predict(X_test)

# Calculate performance metrics for the best XGBoost model
mae_xgb_best = mean_absolute_error(y_test, y_pred_xgb_best)
r2_xgb_best = r2_score(y_test, y_pred_xgb_best)
cv_scores_xgb_best = cross_val_score(best_xgb, X_train, y_train, cv=10, scoring='neg_mean_squared_error')
rmse_scores_xgb_best = np.sqrt(-cv_scores_xgb_best)

# Print results for the best XGBoost model
print("Best XGBoost Regressor")
print("Best Parameters:", grid_search_xgb.best_params_)
print("Mean Absolute Error (MAE):", mae_xgb_best)
print("R2 Score:", r2_xgb_best)
print("Mean RMSE:", rmse_scores_xgb_best.mean())

# Plot true values vs predicted values for best XGBoost model
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_xgb_best, alpha=0.3, color='blue', label='Predicted Values')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, label='Perfect Prediction')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Best XGBoost: True Values vs Predicted Values')
plt.legend()
plt.show()

# Residual Plot for best XGBoost model
plt.figure(figsize=(10, 6))
residuals_xgb = y_test - y_pred_xgb_best
plt.scatter(y_pred_xgb_best, residuals_xgb, alpha=0.3, color='blue')
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Best XGBoost: Residual Plot')
plt.show()

# Plot feature importances for best XGBoost model
feature_importances_xgb = pd.Series(best_xgb.feature_importances_, index=X_train.columns)
top_5_features_xgb = feature_importances_xgb.nlargest(5)


# Print top 5 features
print("Top 5 Features:")
print(top_5_features_xgb)
```
